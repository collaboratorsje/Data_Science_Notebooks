```python
from ISLP import load_data
auto_data = load_data('Auto')
auto_data.head(5)
```

1. Use the sm.OLS() function to perform a multiple linear regression with mpg as the response and all other variables except name as the predictors.

```python
from ISLP.models import ModelSpec as MS
import statsmodels.api as sm
```

```python
x = MS(['cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'year', 'origin']).fit_transform(auto_data) # Set the dependent variables / predictors
y = auto_data['mpg'] # Set the independent variable / response
model_MLR = sm.OLS(y, x) # Create the Multiple Linear Regression Model
result_MLR = model_MLR.fit() # Train the Multiple Linear Regression Model
```

2. Use the summarize() function to print the results.

```python
from ISLP.models import summarize

summarize(result_MLR) # Show the model results
```

3. Fit some models with interactions as described in the lab. Comment on the output - are any interactions statistically significant?

```python
# create an interaction between Horsepower and Weight
x = MS(['horsepower', 'weight', ('horsepower', 'weight')]).fit_transform(auto_data)
model_MLR_HW = sm.OLS(y, x) # Create the Horsepower/Weight interaction model
result_MLR_HW = model_MLR_HW.fit() # Train the Horsepower/Weight interaction model
summarize(result_MLR_HW) # Show the model results
```

```python
# create an interaction between Cylinders and Acceleration that also takes into account all predictors
x = MS(['cylinders', 'acceleration', 'displacement', 'horsepower', 'weight', 'year', 'origin', ('cylinders', 'acceleration')]).fit_transform(auto_data)
model_MLR_CAA = sm.OLS(y, x) # Create the Cylinders/Acceleration interaction model that includes all predictors
result_MLR_CAA = model_MLR_CAA.fit() # Train the Cylinders/Acceleration interaction model that includes all predictors
summarize(result_MLR_CAA) # Show the model results
```

```python
# create an interaction between Cylinders and Acceleration
x = MS(['cylinders', 'acceleration', ('cylinders', 'acceleration')]).fit_transform(auto_data)
model_MLR_CA = sm.OLS(y, x) # Create the Cylinders/Acceleration interaction model that includes all predictors
result_MLR_CA = model_MLR_CA.fit() # Train the Cylinders/Acceleration interaction model that includes all predictors
summarize(result_MLR_CA) # Show the model results
```
